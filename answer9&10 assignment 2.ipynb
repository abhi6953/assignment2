```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome(executable_path='C:/chromedriver/chromedriver.exe')
```

```python
driver.get('https://www.myntra.com/shoes ')
```

```python
price_filter=driver.find_element_by_xpath("//ul[@class='price-list']/li[2]")
price_filter.click()
color_filter=driver.find_element_by_xpath("//li[@class='colour-listItem']/label[1]")
color_filter.click()
```

```python
brand_name=[]
description=[]
price=[]
```

```python
b_tag=driver.find_elements_by_xpath("//h3[@class='product-brand']")
len(b_tag)
```

```python
d_tag=driver.find_elements_by_xpath("//h4[@class='product-product']")
len(d_tag)
```

```python
p_tag=driver.find_elements_by_xpath("//div[@class='product-price']")
len(p_tag)
```

```python
for c in b_tag:
    brand_name.append(c.text)
for c in d_tag:
    description.append(c.text)
for c in p_tag:
    price.append(c.text)
```

```python
shoes=pd.DataFrame({'brand_name':brand_name,'short_description':description,'shoe_price':price})
shoes
```
